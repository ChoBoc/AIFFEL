```python
# 필요한 모듈 import 하기

from sklearn.datasets import load_wine
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report

# 데이터 준비
wine = load_wine()
```

```python
# 데이터 이해하기
wine_data = wine.data
wine_label = wine.target
# target_names 출력해보기
wine.target_names
```

```python
# 데이터 Describe 해보기
print(wine.DESCR)
```

```python
# train, test 데이터 분리
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(wine_data,
                                                   wine_label,
                                                   test_size=0.2,
                                                   random_state=7)
print("X_train 개수: ", len(X_train),", X_test 개수: ", len(X_test))
```

```python
# 1. Decision Tree 사용해보기

from sklearn.tree import DecisionTreeClassifier

wine_tree = DecisionTreeClassifier(random_state=10)
wine_tree.fit(X_train, y_train)
y_pred = wine_tree.predict(X_test)
print(classification_report(y_test, y_pred))

from sklearn.metrics import confusion_matrix
confusion_matrix(y_test, y_pred)
```

```python
# 2. Random Forest 사용해보기

from sklearn.ensemble import RandomForestClassifier

wine_forest = RandomForestClassifier(random_state=10)
wine_forest.fit(X_train, y_train)
y_pred = wine_forest.predict(X_test)
print(classification_report(y_test, y_pred))

from sklearn.metrics import confusion_matrix
confusion_matrix(y_test, y_pred)
```

```python
# 3. SVM 사용해보기

from sklearn import svm
svm_model = svm.SVC()

svm_model.fit(X_train, y_train)
y_pred = svm_model.predict(X_test)
print(classification_report(y_test, y_pred))

from sklearn.metrics import confusion_matrix
confusion_matrix(y_test, y_pred)
```

```python
# 4. SGD Classifier 사용해보기

from sklearn.linear_model import SGDClassifier
sgd_model = SGDClassifier()

sgd_model.fit(X_train, y_train)
y_pred = sgd_model.predict(X_test)
print(classification_report(y_test, y_pred))

from sklearn.metrics import confusion_matrix
confusion_matrix(y_test, y_pred)
```

```python
# 5. Logistic Regression 사용해보기

from sklearn.linear_model import LogisticRegression
logistic_model = LogisticRegression()

logistic_model.fit(X_train, y_train)
y_pred = logistic_model.predict(X_test)
print(classification_report(y_test, y_pred))

from sklearn.metrics import confusion_matrix
confusion_matrix(y_test, y_pred)
```

## 와인 분류 결과

#### 1.0, 100%의 정확도를 낸 RandomForest가 적절한 것 같습니다.

 - 와인 분류 데이터셋에서 특징이 많은 것을 확인했고 이를 의사결정트리를 통해 현명하게 처리해주는 RandomForest 방식이 적절한 평가지표라고 생각합니다.
